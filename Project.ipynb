```python
#Histogram graph of fixed acidity against frequency
plt.hist(dataset['fixed acidity'])
plt.xlabel('fixed acidity')
plt.ylabel('Frequency')
plt.title('Distribution of fixed acidity in red wine')
plt.show()
```

```python
# Scatter diagram of fixed acidity and pH
plt.scatter(dataset['pH'],dataset['fixed acidity'],alpha=0.5)
plt.xlabel('pH')
plt.ylabel('fixed acidity')
plt.title('Relationship between fixed acidity and pH')
plt.show()
```

```python
#Barchart of volatile acidity and quality
x=(dataset['quality'])
y=(dataset['volatile acidity'])
plt.bar(x,y)
plt.xlabel('quality')
plt.ylabel('volatile acidity')
plt.title('Influence of volatile acidity against quality')
plt.show()
```

Fixed acidity is a crucial factor in determining the taste and quality of red wine. It represents the total concentration of acids that do not evaporate. The histogram shows the distribution of values in fixed acidity indicating the frequency of fixed acidity.From histogram distribution it indicates that the mean fixed acidity is about 8.3,its standard deviation is about 1.7, its minimum taste and quality is about 4.6 and maximun taste is about 15.9.
The scatter diagram shows the relationship beween fixed acidity and pH.From the scatter diagram the fixed acidity from 6 - 10 is concentrated from a pH of 3.2 - 3.6. From the scatter diagram it can be deduced that red wines with a pH of about 3.2 - 3.6 have high quality.
Volatile acidity is a crucial factor influencing the flavor of red wine. The bar chart shows the influence of volatile acidity against quality. From the bar chart red wine having a quality of 3 has high sensory characteristics(volatile acidity) followed by 5,4,6,7 and 8 having the least volatile acidity.
